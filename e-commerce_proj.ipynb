### **Final Conclusion**

The analysis reveals that **Set** is the highest revenue-generating category, making it the primary driver of business performance. Geographically, **Maharashtra** emerges as the top contributing state, with a few key cities accounting for a significant portion of total sales, indicating strong regional demand concentration.

The sales trend shows a **fluctuating pattern**, with a clear peak in early May followed by a gradual decline, suggesting the impact of seasonal demand or promotional activities. In terms of operations, **non–Easy Ship (self/unknown fulfillment)** contributes the majority of revenue, indicating a higher reliance on alternative fulfillment methods.

However, a notable percentage of orders are **cancelled (~15%)**, which highlights a critical area for improvement in order management, customer experience, or delivery efficiency. Additionally, **normal (B2C) customers contribute more to overall revenue compared to B2B**, showing that the business primarily depends on individual buyers rather than bulk or business orders.

### Overall Insight:
The business demonstrates strong performance in specific product categories and regions but relies heavily on a limited set of drivers. To achieve sustainable growth, the focus should be on reducing cancellations, optimizing fulfillment strategies, and expanding performance in underutilized categories and regions.
